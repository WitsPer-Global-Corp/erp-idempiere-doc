````python
import os
from googletrans import Translator

translator = Translator()

input_dir = 'D:\Code\idempiere-doc\erp-idempiere-doc\docs'      #  md 
output_dir = 'D:\Code\idempiere-doc\erp-idempiere-doc\docs-zh'  # 

input_dir = 'D:\Code\idempiere-doc\erp-idempiere-doc\docs\\basic-development'      #  md 
output_dir = 'D:\Code\idempiere-doc\erp-idempiere-doc\docs-zh\\basic-development'  # 

input_dir = 'D:\Code\idempiere-doc\erp-idempiere-doc\docs\\basic-functional'      #  md 
output_dir = 'D:\Code\idempiere-doc\erp-idempiere-doc\docs-zh\\basic-functional'  # 

input_dir = 'D:\Code\idempiere-doc\erp-idempiere-doc\docs\\basic-installation'      #  md 
output_dir = 'D:\Code\idempiere-doc\erp-idempiere-doc\docs-zh\\basic-installation'  # 



input_dir = 'D:\Code\idempiere-doc\erp-idempiere-doc\docs-en\\basic-installation\manual-install'      #  md 
output_dir = 'D:\Code\idempiere-doc\erp-idempiere-doc\docs\\basic-installation\manual-install'  # 

# input_dir = 'D:\Code\idempiere-doc\erp-idempiere-doc\docs\\basic-primers'      #  md 
# output_dir = 'D:\Code\idempiere-doc\erp-idempiere-doc\docs-zh\\basic-primers'  # 

os.makedirs(output_dir, exist_ok=True)


def translate_md_content(content):
    translated_lines = []
    code_block = False
    yaml_block = False
    buffer = []

    lines = content.splitlines(keepends=True)  # 

    for line in lines:
        stripped = line.strip()

        #  YAML frontmatter ( ---)
        if stripped == "---":
            yaml_block = not yaml_block
            translated_lines.append(line)
            continue

        # 
        if stripped.startswith("```"):
            code_block = not code_block
            translated_lines.append(line)
            continue

        # 
        if stripped == "":
            translated_lines.append(line)
            continue

        #  YAML  code 
        if yaml_block or code_block:
            translated_lines.append(line)
            continue

        # 
        prefix = ""
        if line.lstrip().startswith(("* ", "- ", "+ ", "> ", "#")):
            parts = line.lstrip().split(" ", 1)
            if len(parts) == 2:
                prefix, rest = parts
                try:
                    translated = translator.translate(rest, src='en', dest='zh-tw').text
                    translated_lines.append(line[:len(line)-len(line.lstrip())] + prefix + " " + translated + "\n")
                except Exception as e:
                    print(f" : {e}")
                    translated_lines.append(line)
            else:
                translated_lines.append(line)
            continue

        # 
        try:
            translated = translator.translate(stripped, src='en', dest='zh-tw').text
            translated_lines.append(translated + "\n")
        except Exception as e:
            print(f" : {e}")
            translated_lines.append(line)

    return "".join(translated_lines)

#  .md 
for filename in os.listdir(input_dir):
    if filename.endswith('.md'):
        input_path = os.path.join(input_dir, filename)
        output_path = os.path.join(output_dir, filename)

        with open(input_path, 'r', encoding='utf-8') as f:
            content = f.read()

        translated_content = translate_md_content(content)

        with open(output_path, 'w', encoding='utf-8') as f:
            f.write(translated_content)

        print(f" {filename}")
print(f" ")
````
